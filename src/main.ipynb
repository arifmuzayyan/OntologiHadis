```python
df_scraper['Terjemahan'] = df_scraper['Terjemahan'].str.replace("; \"",": \"")
df_scraper['Matan'] = [x.split(": \"",1)[-1] for x in df_scraper.Terjemahan]
df_scraper.head()
```

## Preproccess

```python
# df_bukhari = df_sunnah
# df_bukhari = df_matan
df_bukhari = df_scraper
```

#### Remove punctuation

```python
import string
# df_bukhari['Praproses'] = df_bukhari['Terjemahan'].str.replace('[{}]'.format(string.punctuation), '', regex=True)
df_bukhari['Praproses'] = df_bukhari['Matan'].str.replace('[{}]'.format(string.punctuation), '', regex=True)
df_bukhari.head()
```

#### Case folding

```python
df_bukhari['Praproses'] = df_bukhari['Praproses'].str.lower()
df_bukhari.head()
```

#### Stopwords removal

```python
# !pip install Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()
stopwords = factory.create_stop_word_remover()

def stopwords_wrapper(text):
    return stopwords.remove(text)
df_bukhari['Praproses'] = df_bukhari['Praproses'].apply(stopwords_wrapper)
df_bukhari.head()
```

#### Tokenization

```python
# !pip install nltk
# import nltk
# nltk.download()
```

```python
from nltk.tokenize import word_tokenize 

def word_tokenize_wrapper(text):
    return word_tokenize(text)

df_bukhari['Token'] = df_bukhari['Praproses'].apply(word_tokenize_wrapper)
df_bukhari.head()
```

#### POS tagging

```python
# !pip install numpy
# !pip install python-crfsuite
```

```python
from nltk.tag import CRFTagger
ct = CRFTagger()
ct.set_model_file('all_indo_man_tag_corpus_model.crf.tagger')
df_bukhari['POS tag'] = ct.tag_sents(df_bukhari['Token'])
df_bukhari['POS tag'].loc[0]
```

## Conceptualization

#### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=250)
tfidf = vectorizer.fit_transform(df_bukhari['Praproses'])

df_ontologi = pd.DataFrame(vectorizer.get_feature_names_out(), columns=['konsep'])

print(df_ontologi['konsep'])
print("TF-IDF Matrix: ")
print(tfidf.toarray())
```

#### Word embedding

```python
# !pip install gensim
```

```python
from gensim.models import Word2Vec

skipgram_model = Word2Vec(sentences=df_bukhari['Token'],
                          vector_size=200,  # Dimensionality of the word vectors
                          window=5,         # Maximum distance between the current and predicted word within a sentence
                          sg=1,             # Skip-Gram model (1 for Skip-Gram, 0 for CBOW)
                          min_count=20,      # Ignores all words with a total frequency lower than this
                          workers=4)       # Number of CPU cores to use for training the model
 
# Training
skipgram_model.train(df_bukhari['Token'], total_examples=1, epochs=10)
skipgram_model.save("skipgram_model.model")
```

```python
def word_embedding(text):
    loaded_model = Word2Vec.load("skipgram_model.model")
    return(loaded_model.wv[text])

df_ontologi['relasi'] = df_ontologi['konsep'].apply(word_embedding)
df_ontologi
```

#### Agglomerative hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np

X = np.array(df_ontologi['relasi'].to_list())

clustering = AgglomerativeClustering(n_clusters=None,
                                     metric='euclidean',
                                     linkage='average',
                                     distance_threshold=2.73
                                     ).fit(X)

print(clustering.labels_)
df_ontologi['cluster'] = clustering.labels_
df_ontologi.head()
```

```python
ontologi = df_ontologi.groupby('label')
ontologi.apply(display)
```

## Serialization

```python
from rdflib.namespace import OWL, RDF, RDFS
from rdflib import Graph, Literal, Namespace, URIRef

# Construct the hadis namespace
Hadis = Namespace("http://www.hadits.id/#")

# Create the graph
g = Graph()

# Bind the OWL and Hadis name spaces
g.bind("owl", OWL)
g.bind("hadis", Hadis)

g.add((URIRef('http://www.semanticweb.org/habib/ontologies/2025/2/shahih_bukhari.owl'), RDF.type, OWL.Ontology ))

# Create the node to add to the Graph
Bukhari = URIRef(Hadis["shahih_bukhari"])

# Add the OWL data to the g
g.add((Bukhari, RDF.type, OWL.Class))
g.add((Bukhari, RDFS.subClassOf, OWL.Thing))
g.add((Bukhari, RDFS.label, Literal("Shahih Bukhari")))
g.add((Bukhari, RDFS.comment, Literal("The ontology of the hadith shahih bukhari")))

for cluster in ontologi:

    Konsep = URIRef(Bukhari+str(cluster[0]))
    g.add((Konsep, RDF.type, OWL.Class))
    g.add((Konsep, RDFS.subClassOf, Bukhari))
    g.add((Konsep, RDFS.label, Literal(cluster[0])))

    for daun in cluster[1]['konsep']:

        Daun = URIRef(Konsep+daun)
        g.add((Daun, RDF.type, OWL.Class))
        g.add((Daun, RDFS.subClassOf, Konsep))
        g.add((Daun, RDFS.label, Literal(daun)))

print(g.serialize('shahih_bukhari.owl', format='pretty-xml'))
# print(g.serialize(format='pretty-xml'))
```
