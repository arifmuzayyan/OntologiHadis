```python
import pandas as pd

df_hadis = pd.read_csv("Sunnah.csv")
df_hadis
```

```python
df_hadis = df_hadis.dropna(how='any',axis=0)
df_hadis = df_hadis.drop('Arab', axis=1)
df_hadis
```

```python
df_bukhari = df_hadis.loc[8724:15361].reset_index()
df_bukhari
```

## Preproccess

#### Remove punctuation

```python
import string
df_bukhari['Praproses'] = df_bukhari['Terjemahan'].str.replace('[{}]'.format(string.punctuation), '', regex=True)
df_bukhari.head()
```

#### Case folding

```python
df_bukhari['Praproses'] = df_bukhari['Praproses'].str.lower()
df_bukhari.head()
```

#### Stopwords removal

```python
# !pip install Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()
stopwords = factory.create_stop_word_remover()

def stopwords_wrapper(text):
    return stopwords.remove(text)
df_bukhari['Praproses'] = df_bukhari['Praproses'].apply(stopwords_wrapper)
df_bukhari.head()
```

#### Tokenization

```python
# !pip install nltk
# import nltk
# nltk.download()
```

```python
from nltk.tokenize import word_tokenize 

def word_tokenize_wrapper(text):
    return word_tokenize(text)

df_bukhari['Token'] = df_bukhari['Praproses'].apply(word_tokenize_wrapper)
df_bukhari.head()
```

#### POS tagging

```python
# !pip install numpy
# !pip install python-crfsuite
```

```python
from nltk.tag import CRFTagger
ct = CRFTagger()
ct.set_model_file('all_indo_man_tag_corpus_model.crf.tagger')
df_bukhari['POS tag'] = ct.tag_sents(df_bukhari['Token'])
df_bukhari['POS tag'].loc[0]
```
